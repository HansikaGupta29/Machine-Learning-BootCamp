```python
l= train.shape[1]

X_train= train.iloc[:, 1:l-1]
y_train= train.iloc[:, 0]

X_train= X_train.to_numpy()
y_train= y_train.to_numpy()
y_train= y_train.reshape(X_train.shape[0], 1)
print(X_train.shape)
y_train.shape
```

```python
def sigmoid(z):
  g= 1/(1+np.exp(-1*z))
  return g
```

```python
m= X_train.shape[0]

X_train= np.c_[np.ones(m), X_train]
n= X_train.shape[1]
print(n)
```

```python
def Y_edit(Y):
  Y_edited= np.zeros((Y.shape[0], Y.max() ))
  Y_edited[np.arange(Y.shape[0]), Y-1 ]= 1
  return Y_edited
```

```python
def costFunction(X, y, theta, lamda):
  m= y.shape[0]
  J=0
  grad= np.zeros((theta.shape))
  temp= np.zeros((theta.shape))

  predictions= sigmoid(np.dot(X, theta))
  
  value= y* np.log(predictions + 10**-7) + (1-y)* np.log((1-predictions) + 10**-7)
  J=-1/m * np.sum(value) + lamda/(2*m) * (np.sum(theta**2) - theta[1]**2)

  return J
```

```python
def gradientDescent(X_train, y_train, theta, alpha, num_iters, lamda):
  n= X_train.shape[1]
  print(n)
  J_history= np.zeros((num_iters, k))
  

  for iter in range(0, num_iters-1):
    grad= np.zeros((n, k))
    predictions= sigmoid(np.dot(X_train, theta))

    
    for i in range(1, n):
      if i==0:
        grad[i, :]= np.sum(np.dot(X_train.T, predictions - y_train))/m
      else:
        grad[i, :]= np.sum(np.dot(X_train.T, predictions - y_train))/m + lamda/m * theta[i, 0]

    theta= theta - alpha*grad
    J_history[iter, :]= costFunction(X_train, y_train, theta, lamda)

  return [theta, J_history]
```

```python
alpha = 0.00001
num_iters= 1500
lamda= 1
k= y_train.max()

theta= np.zeros((n, k))

y_train= Y_edit(y_train)

J_hist= np.zeros((num_iters, k))
[theta, J_hist]= gradientDescent(X_train, y_train, theta, alpha, num_iters, lamda)
print("Theta calculated from gradient descent is ", theta)
theta.shape
```

```python
print(J_hist)
plt.plot(range(num_iters), J_hist, 'bx')
```
