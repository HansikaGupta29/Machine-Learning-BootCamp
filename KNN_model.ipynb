```python
train = pd.read_csv('/content/drive/MyDrive/Cyberlabs- ML/EMNIST dataset/emnist-letters-train.csv')
train.head()
```

```python
test = pd.read_csv('/content/drive/MyDrive/Cyberlabs- ML/EMNIST dataset/emnist-letters-test.csv')
test.head()
```

```python
l_train= train.shape[1]

X_train= train.iloc[:, 1:l_train-1]
y_train= train.iloc[:, 0]

X_train= X_train.to_numpy()
y_train= y_train.to_numpy()
y_train= y_train.reshape(X_train.shape[0], 1)
print(X_train.shape)
y_train.shape

m_train= X_train.shape[0]
```

```python
l_test= test.shape[1]

X_test= test.iloc[:, 1:l_test-1]
y_test= test.iloc[:, 0]

X_test= X_test.to_numpy()
y_test= y_test.to_numpy()
y_test= y_test.reshape(X_test.shape[0], 1)
print(X_test.shape)
y_test.shape

m_test= X_test.shape[0]
```

```python
def euclidean_distance(pt1, pt2):
  dis= np.sqrt(np.sum((pt1-pt2)**2))
  return dis
```

```python
def KNN( H, k_val ): 
  distance=[] 
  y_k_neighbours= []
  for i in range(m_train): 
    distance.append(euclidean_distance(H, X_train[i, :])) 
  k_neighbours= np.argsort(distance)[:k_val]
  
  for j in range(k_val): 
    y_k_neighbours.append(y_train[k_neighbours[j]][0])

  return max(set(y_k_neighbours), key=y_k_neighbours.count)
```

```python
k=7
predict=[]

for j in range(m_test):
  predict.append(KNN(X_test[j,:], k))
```

```python
count= 0

for i in range(m_test):
  if predict[i]== y_test[i][0]:
    count+=1

accuracy= count/m_test * 100

print(accuracy)
```
